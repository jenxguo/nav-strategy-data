```python
import pandas as pd
from IPython.display import display
from collections import namedtuple

txtFiles = [
    '204_sex0.txt',
    '204a_sex0.txt',
    '205_sex1.txt',
    '205a_sex1.txt',
    '206_sex0.txt',
    '206a_sex0.txt',
    '207_sex1.txt',
    '207a_sex1.txt',
    '208_sex1.txt',
    '208a_sex1.txt',
    '209_sex0.txt',
    '209a_sex0.txt',
    '210_sex1.txt',
    '210a_sex1.txt',
    '211_sex1.txt',
    '211a_sex1.txt',
    '212_sex0.txt',
    '212a_sex0.txt',
    '214_sex0.txt',
    '214a_sex0.txt',
    '215_sex1.txt',
    '215a_sex1.txt',
    '216_sex0.txt',
    '216a_sex0.txt',
    '217_sex1.txt',
    '217a_sex1.txt',
    '218_sex0.txt',
    '218a_sex0.txt',
    '220_sex1.txt',
    '220a_sex1.txt',
    '221_sex0.txt',
    '221a_sex0.txt',
    '222_sex1.txt',
    '222a_sex1.txt',
    '223_sex1.txt',
    '223a_sex1.txt',
    '224_sex0.txt',
    '224a_sex0.txt',
    '225_sex1.txt',
    '225a_sex1.txt',
    '227_sex1.txt',
    '227a_sex1.txt',
    '228_sex0.txt',
    '228a_sex0.txt',
    '229_sex0.txt',
    '229b_sex0.txt',
    '230_sex1.txt',
    '230b_sex1.txt',
    '231_sex0.txt',
    '231a_sex0.txt',
    '232_sex1.txt',
    '232a_sex1.txt',
    '233_sex0.txt',
    '233b_sex0.txt',
    '234_sex0.txt',
    '234b_sex0.txt',
    '236_sex0.txt',
    '236a_sex0.txt',
    '237_sex1.txt',
    '237a_sex1.txt',
    '238_sex0.txt',
    '238a_sex0.txt',
    '239_sex1.txt',
    '239b_sex1.txt',
]

Item = namedtuple('Item', 'ParticipantNum ParticipantGender Date ExperimentType AltExperiment Time XPosition YPosition Orientation BoxLocation')
items = []

for file in txtFiles:
    lineNum = 0
    participantNum = ''
    participantGender = ''
    date = ''
    experimentType = ''
    altExperiment = ''
    with open(file) as f:
        for line in f:
            line = line.rstrip('\n')
            if lineNum > 7:
                s = line.split(':  ')
                if len(s) > 1:
                    time = s[0]
                    p = s[1].split(',')
                    xPos = p[0]
                    yPos = p[1]
                    orientation = p[2]
                    boxLoc = p[3]
                    items.append(Item(participantNum, participantGender, date, experimentType, altExperiment, time, xPos, yPos, orientation, boxLoc))
            else:
                if lineNum == 1:
                    s = line.split(': ')
                    participantNum = s[1]
                if lineNum == 2:
                    s = line.split(': ')
                    if s[1] == 0:
                        participantGender = 'Male'
                    else:
                        participantGender = 'Female'
                if lineNum == 3:
                    s = line.split(': ')
                    date = s[1]
                if lineNum == 4:
                    s = line.split(': ')
                    experimentType = s[1]
                if lineNum == 5:
                    s = line.split(': ')
                    altExperiment = s[1]
            lineNum += 1

df = pd.DataFrame.from_records(items, columns=['Participant No.', 'Participant Gender', 'Date', 'Session', 'Alt Experiment', 'Time', 'X Position', 'Y Position', 'Orientation', 'Box Location'])
display(df)

df.to_csv('CleanedExperiment1.csv')
```
